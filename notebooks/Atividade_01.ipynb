```python
# Importar bibliotecas
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# Ler base de dados (atualize o caminho para a pasta do seu drive (no drive clique com o botão direito e 'copiar caminho'))
tips = pd.read_csv('../data/tips.csv')
```

```python
# Ler base de dados (importar do banco de dados embutido no seaborn)
tips = sns.load_dataset('tips')
tips_df = tips.copy()
tips_df.sample(10)
```

```python
# Informação geral e tipos de dados
tips.info()
```

```python
# Estatísticas descritivas - dados numéricos
tips.describe()
```

```python
# Estatísticas descritivas - dados categóricos
tips[['sex', 'smoker', 'day', 'time']].describe()
```

```python
# Plotar histograma
tips['Bill Amount'].hist()
print(tips['Bill Amount'].agg([np.mean, np.median]))
plt.show()
```

```python
plt.figure(figsize=(10, 8))
sns.set_theme(style='darkgrid')
ax = sns.histplot(data = tips, x='Bill Amount', kde=True)
ax.set_title('Bill Amount Histogram')
plt.show()
```

```python
plt.figure(figsize=(10, 8))
sns.set_theme(style='darkgrid')
ax = sns.histplot(data = tips, x='Tip Amount', kde=True)
ax.set_title('Tip Amount Histogram')
plt.show()
```

## Countplot 
Mostra a contagem de observações em cada categoria observada.
Pode ser visto como um histograma que se aplica a variáveis categóricas (qualitativas), em vez de numéricas (quantitativas).
Permite comparar contagens de variáveis aninhadas.

```python
plt.figure(figsize=(10, 8))
sns.set_theme(style='darkgrid')
ax = sns.countplot(data = tips, x='Day of Week')
plt.show()
```

```python
plt.figure(figsize=(10, 8))
sns.set_theme(style='darkgrid')
ax = sns.countplot(data = tips, x='Server')
plt.show()
```

```python
plt.figure(figsize=(10, 8))
sns.set_theme(style='darkgrid')
ax = sns.countplot(data = tips, x='Day of Week', hue='Server')
plt.show()
```

```python
plt.figure(figsize=(10, 8))
sns.set_theme(style='darkgrid')
ax = sns.barplot(data = tips, x='Server', y='Tip Amount')
plt.show()
```

## Barplot
Representa uma estimativa de tendência central para uma variável numérica com a altura de cada retângulo e fornece um indicador de incerteza dessa estimativa usando barras de erros.

```python
plt.figure(figsize=(10, 8))
sns.set_theme(style='darkgrid')
ax = sns.barplot(data = tips, x='Day of Week', y='Tip Amount')
plt.show()
```

```python
plt.figure(figsize=(10, 8))
sns.set_theme(style='darkgrid')
ax = sns.scatterplot(data = tips, x='Bill Amount', y='Tip Amount', hue='Day of Week', alpha=0.5)
plt.show()
```

```python
plt.figure(figsize=(10, 8))
sns.set_theme(style='darkgrid')
ax = sns.relplot(data = tips, x='Bill Amount', y='Tip Amount', hue='Day of Week', col='Server', alpha=0.5)
plt.show()
```

```python
plt.figure(figsize=(10, 8))
sns.set_theme(style='darkgrid')
ax = sns.relplot(data = tips, x='Bill Amount', y='Tip Amount', hue='Server', col='Day of Week')
plt.show()
```
